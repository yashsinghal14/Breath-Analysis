```python
# Full end-to-end pipeline (copy & run)
import os
import joblib
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.feature_selection import SelectFromModel, VarianceThreshold, SelectKBest, mutual_info_classif
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
from sklearn.inspection import permutation_importance
import warnings
warnings.filterwarnings("ignore")

# ---------- CONFIG ----------
TRAIN = "/content/test_set_fu.csv"
TEST  = "/content/test_set_fu.csv"
TARGET_GUESS = "label"   # pipeline will fallback to last column if not found
N_FOLDS = 5
RANDOM_STATE = 42
# Feature selection target dimension (tweakable): choose top_k features (set None to skip SelectKBest)
TOP_K = 200   # 200 is a reasonable reduction for 1620 features. Lower if you need faster runs.
# ----------------------------

# 1. Load
train = pd.read_csv(TRAIN)
test = pd.read_csv(TEST)

# 2. Infer target
if TARGET_GUESS in train.columns:
    target_col = TARGET_GUESS
else:
    target_col = train.columns[-1]
print("Inferred target:", target_col)

X_train = train.drop(columns=[target_col])
y_train = train[target_col]
if target_col in test.columns:
    X_test = test.drop(columns=[target_col])
    y_test = test[target_col]
else:
    X_test = test.copy()
    y_test = None

print("Shapes: X_train", X_train.shape, "X_test", X_test.shape, "y_train", y_train.shape)

# 3. Quick EDA prints
print("Target value counts:\n", y_train.value_counts())

# 4. Identify numeric & categorical columns
num_cols = X_train.select_dtypes(include=['number']).columns.tolist()
cat_cols = X_train.select_dtypes(include=['object','category','bool']).columns.tolist()
print("Num cols:", len(num_cols), "Cat cols:", len(cat_cols))

# 5. Drop constant features (fast)
vt = VarianceThreshold(threshold=0.0)
vt.fit(X_train[num_cols])
keep_mask = vt.get_support()
num_kept = np.sum(keep_mask)
kept_num_cols = [c for i,c in enumerate(num_cols) if keep_mask[i]]
print(f"Dropped {len(num_cols)-num_kept} constant numeric cols; kept {num_kept}")

# 6. Preprocessing pipelines
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

transformers = []
if kept_num_cols:
    transformers.append(('num', num_pipeline, kept_num_cols))
if cat_cols:
    transformers.append(('cat', cat_pipeline, cat_cols))

from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer(transformers=transformers, remainder='drop', verbose_feature_names_out=False)

# 7. Optional feature selection AFTER preprocessing:
# We'll use a fast filter (SelectKBest with mutual_info_classif) to reduce to TOP_K features
selector = None
if TOP_K is not None:
    selector = SelectKBest(mutual_info_classif, k=min(TOP_K, len(kept_num_cols) + sum(X_train[c].nunique() for c in cat_cols)))

# 8. Define candidate models (3 models)
models = {
    "LogisticRegression": Pipeline([
        ('pre', preprocessor),
        ('sel', selector) if selector is not None else ('passthrough', 'passthrough'),
        ('clf', LogisticRegression(penalty='l1', solver='saga', C=1.0, max_iter=2000, random_state=RANDOM_STATE))
    ]),
    "RandomForest": Pipeline([
        ('pre', preprocessor),
        ('sel', selector) if selector is not None else ('passthrough', 'passthrough'),
        ('clf', RandomForestClassifier(n_estimators=100, max_depth=20, n_jobs=-1, random_state=RANDOM_STATE))
    ]),
    "HistGradientBoosting": Pipeline([
        ('pre', preprocessor),
        ('sel', selector) if selector is not None else ('passthrough', 'passthrough'),
        ('clf', HistGradientBoostingClassifier(max_iter=200, random_state=RANDOM_STATE))
    ])
}

# 9. Cross-validate each model
cv = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=RANDOM_STATE)
scoring = {'accuracy':'accuracy', 'f1_macro':'f1_macro'}
cv_results = {}
for name, pipe in models.items():
    print("CV start:", name)
    res = cross_validate(pipe, X_train, y_train, cv=cv, scoring=scoring, return_train_score=False, n_jobs=1)
    acc = np.mean(res['test_accuracy'])
    f1m = np.mean(res['test_f1_macro'])
    cv_results[name] = {'accuracy': acc, 'f1_macro': f1m, 'raw': res}
    print(f"  {name}  -> accuracy: {acc:.4f}  f1_macro: {f1m:.4f}")

# 10. Choose best by primary metric = accuracy (for multiclass)
best_model_name = max(cv_results.keys(), key=lambda k: cv_results[k]['accuracy'])
print("Best model:", best_model_name, cv_results[best_model_name])

# 11. Fit best model on full train and evaluate on test (if available)
best_pipe = models[best_model_name]
best_pipe.fit(X_train, y_train)

if y_test is not None:
    y_pred = best_pipe.predict(X_test)
    acc_test = accuracy_score(y_test, y_pred)
    f1m_test = f1_score(y_test, y_pred, average='macro')
    print("Test accuracy:", acc_test, "Test f1_macro:", f1m_test)

# 12. Feature importance (approx)
# Try to use coef_ or feature_importances_ if available, else permutation importance.
feature_names = None
# derive feature names from preprocessor
try:
    # get names after preprocessing
    pre = best_pipe.named_steps['pre']
    feature_names = pre.get_feature_names_out(input_features=kept_num_cols + cat_cols)
except Exception:
    # fallback
    feature_names = kept_num_cols + cat_cols

est = best_pipe.named_steps['clf']
fi = None
if hasattr(est, 'feature_importances_'):
    fi = pd.Series(est.feature_importances_, index=feature_names).sort_values(ascending=False)
elif hasattr(est, 'coef_'):
    coef = est.coef_
    if coef.ndim > 1:
        coef = np.mean(np.abs(coef), axis=0)
    else:
        coef = np.abs(coef)

    # Adjust feature names if feature selection reduced them
    if 'sel' in best_pipe.named_steps and best_pipe.named_steps['sel'] != 'passthrough':
        try:
            mask = best_pipe.named_steps['sel'].get_support()
            reduced_names = np.array(feature_names)[mask]
        except Exception:
            # fallback if mask not available
            reduced_names = feature_names[:len(coef)]
        feature_names_final = reduced_names
    else:
        feature_names_final = feature_names[:len(coef)]

    fi = pd.Series(coef, index=feature_names_final).sort_values(ascending=False)
else:
    # permutation importance (slower)
    print("Computing permutation importance (may take some time)...")
    r = permutation_importance(best_pipe, X_train, y_train, n_repeats=10, random_state=RANDOM_STATE, n_jobs=1)
    fi = pd.Series(r.importances_mean, index=feature_names).sort_values(ascending=False)

print("Top 30 features:")
print(fi.head(30))

# 13. Save the final model
OUT_PATH = "/content/best_model.pkl"
joblib.dump(best_pipe, OUT_PATH)
print("Saved best model to:", OUT_PATH)

# 14. Save CV summary to CSV
cv_summary_df = pd.DataFrame([
    {'model': name, 'cv_accuracy': cv_results[name]['accuracy'], 'cv_f1_macro': cv_results[name]['f1_macro']}
    for name in cv_results
]).sort_values('cv_accuracy', ascending=False)
cv_summary_df.to_csv("/content/cv_summary.csv", index=False)
print("Saved CV summary to: /content/cv_summary.csv")
```

```python
import streamlit as st
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
import sys, types

# Fix numpy._core issue
import numpy as np
if not hasattr(np, "_core"):
    np._core = types.SimpleNamespace()
    sys.modules['numpy._core'] = np.core

# --- Load model and extract feature importances ---
MODEL_PATH = "best_model.pkl"
CV_PATH = "cv_summary.csv"

@st.cache_resource
def load_model():
    model = joblib.load(MODEL_PATH)
    return model

@st.cache_data
def load_cv_summary():
    return pd.read_csv(CV_PATH)

st.set_page_config(page_title="Model Feature Importance Dashboard", layout="wide")
st.title("üìä Model Feature Importance Dashboard")

# Load model and data
model = load_model()
cv_summary = load_cv_summary()
st.subheader("Cross-Validation Results")
st.dataframe(cv_summary)

# Extract feature importance
feature_names = []
try:
    pre = model.named_steps['pre']
    feature_names = pre.get_feature_names_out()
except Exception:
    feature_names = []

estimator = model.named_steps['clf']

if hasattr(estimator, "coef_"):
    coefs = estimator.coef_
    if coefs.ndim > 1:
        coefs = np.mean(np.abs(coefs), axis=0)
    else:
        coefs = np.abs(coefs)
    if 'sel' in model.named_steps and model.named_steps['sel'] != 'passthrough':
        mask = model.named_steps['sel'].get_support()
        reduced_names = np.array(feature_names)[mask]
    else:
        reduced_names = feature_names[:len(coefs)]
    fi = pd.Series(coefs, index=reduced_names).sort_values(ascending=False)
elif hasattr(estimator, "feature_importances_"):
    fi = pd.Series(estimator.feature_importances_, index=feature_names).sort_values(ascending=False)
else:
    fi = None

# --- Visualization ---
if fi is not None and not fi.empty:
    st.subheader("üîù Top 20 Most Important Features")
    top_features = fi.head(20)
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.barh(top_features.index[::-1], top_features.values[::-1], color="skyblue")
    ax.set_xlabel("Importance (absolute coefficient value)")
    ax.set_ylabel("Feature Name")
    ax.set_title("Top 20 Most Important Features (Model Interpretation)")
    st.pyplot(fig)

    # Add table
    st.subheader("üìã Full Feature Importance Table")
    st.dataframe(fi.reset_index().rename(columns={'index': 'Feature', 0: 'Importance'}))
else:
    st.warning("No feature importance data available.")

st.markdown("---")
st.caption("Developed with ‚ù§Ô∏è using Streamlit and scikit-learn.")
```
